```python
#Randomly picks 50 numbers
x = np.random.rand(50)
y = np.random.rand(50)
```

Scatter Plot : randomly scattered

```python
plt.scatter(x,y)
```

We can control shape,size,color,etc

```python
# Alpha - controls the intensity of the color
#marker - representation of the points
plt.scatter(x,y,color='#1E4B37',alpha=.8,marker='x')
plt.xlabel("This is my x axis")
plt.ylabel("This is my y-axis")
plt.title("This is my scatter plot")
```

Grid

```python
plt.scatter(x,y,color='#1E4B37',alpha=.8,marker='x')
plt.xlabel("This is my x axis")
plt.ylabel("This is my y-axis")
plt.title("This is my scatter plot")
plt.grid()
```

Figure size

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y,color='#1E4B37',alpha=.8,marker='x')
plt.xlabel("This is my x axis")
plt.ylabel("This is my y-axis")
plt.title("This is my scatter plot")
plt.grid()
```

```python
data = [1,2,3,4,4,5,5,6,6,7,8,8,8]
```

Histogram(Frequency Mapping) : gives Unique nums itself on x -axis and frequency on y-axis

```python
plt.hist(data,color='#700EBE')
plt.xlabel("Unique Values")
plt.ylabel("Frequency")
plt.title("Frequency Mapping")
plt.grid()
```

To save the graph

```python
plt.hist(data,color='#700EBE')
plt.xlabel("Unique Values")
plt.ylabel("Frequency")
plt.title("Frequency Mapping")
plt.grid()
plt.savefig("test.png") #graph saved as test.png
```

3D Graph

```python
x = np.random.rand(50)
y= np.random.rand(50)
z = np.random.rand(50)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z,color='m',alpha=.9)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# plt.zlabel("Z-axis")
plt.show()
```
