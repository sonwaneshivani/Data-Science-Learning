Merge and Join Operations

```python
data1 = pd.DataFrame({"Key1":(1,2,4,5,6),
     "Key2": (4,5,6,7,8),
     "Key3" : (3,4,5,6,6)})
```

```python
data2 = pd.DataFrame({"Key1":(1,2,45,6,67),
     "Key4": (56,5,6,7,8),
     "Key5" : (3,56,5,6,6)})
```

```python
data1
```

```python
data2
```

It is very similar to venn diagram. These are useful in data filteration and data combining
* Only merge the common ones(intersection), Rest are ignored - This is how=inner(by default)

```python
pd.merge(data1,data2)
```

how = 'left'.., Here it will try to keep everything(all columns) from left, and try to keep all distinct columns from right (only wrto the available left side(common ones) rest are ignored and placed NaN)

```python
pd.merge(data1,data2,how='left')
```

how='Right'.., Viceversa of left

```python
pd.merge(data1,data2,how='right')
```

Outer : Brings everything, all the combinations

```python
pd.merge(data1,data2,how='outer')
```

Cross merge creates combinations of all the possible outcomes. It picks one row and combines with all other from another dataframe

```python
pd.merge(data1,data2,how='cross')
```

Purposively applying the operation on particular column

```python
pd.merge(data1,data2,how='left')
```

```python
pd.merge(data1,data2,how='left',on="Key1")
```

```python
data1
```

```python
data2
```

Here it will check wrto key2 with key4

```python
pd.merge(data1,data2,how='left',left_on="Key2", right_on="Key4")
```

Join

```python
data1 = pd.DataFrame({
    'Key1':[1,2,4,5,6],
    'Key2':[4,5,6,7,8],
    'Key3':[3,4,5,6,6]},
    index=['a','b','c','d','e']
)
```

```python
data2 = pd.DataFrame({
    'Key11':[1,2,4,5,6],
    'Key22':[4,5,6,7,8],
    'Key33':[3,4,5,6,6]},
    index=['a','b','h','i','j']
)
```

```python
data1
```

```python
data2
```

Join - It tries to match indexes(not columns)
Merge - It tries to match  columns

```python
#By default how - left
data1.join(data2)
```

```python
#Common
data1.join(data2,how="inner")
```

```python
data1.join(data2,how="right")
```

```python
#Left and right
data1.join(data2,how="outer")
```

```python
#Combinations
data1.join(data2,how="cross")
```
