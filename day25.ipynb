Difference btwn iloc(Integer location) & loc(location func)
* Loc always try to consider named indexes. Considers upper range
* iloc always go  with inbuilt ones doesn't consider upper bound

Trying to consider rows as well as columns:
* Here we didn't get error with loc but got an error with iloc.
* Iloc goes with the internal/default indexes given by the system
* Loc understands whatever we have like named ones

```python
df.iloc[0:2,["PassengerId",	"Survived","Pclass"]]
```

```python
df.loc[0:2,["PassengerId",	"Survived","Pclass"]]
```
